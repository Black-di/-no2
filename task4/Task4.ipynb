## 获取Cookies
由于验证码还没有调通，先使用Cookie方式。
首先通过selenium模拟登录并获取Cookies信息

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import json

browser = webdriver.Chrome()
browser.get("http://www.dxy.cn/bbs/thread/626626#626626")
login = browser.find_element_by_class_name('nav_account')
login.click()
login = browser.find_element_by_xpath('//html/body/div[2]/div[2]/div[1]/a[2]')
login.click()
inputbox = browser.find_element_by_css_selector('#username')
inputbox.send_keys('********')
inputbox = browser.find_element_by_xpath('//*[@id="user"]/div[1]/div[1]/div[1]/div[2]/input')
inputbox.send_keys('********')
button = browser.find_element_by_css_selector('#user > div.login__content > div.form__button > button')
button.click()
time.sleep(10)     #这里等待10秒，通过人工输入验证码
cookies = browser.get_cookies()
with open("cookies.txt", "w") as fp:    #将回去到得值写入本地文本
    json.dump(cookies, fp)
```

## 使用Cookies登录并获取ID和评论

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome()
browser.get("http://www.dxy.cn/bbs/thread/626626#626626")

with open("cookies.txt", "r") as fp:
        cookies = json.load(fp)
        for cookie in cookies:
            if 'expiry' in cookie:   #测试发现添加'expiry'会报错，因此在这将其跳过
                continue
            browser.add_cookie(cookie)
browser.refresh()

soup = BeautifulSoup(browser.page_source,'lxml')
posts = soup.find(attrs={'class':'postbox'}).children  #这种做法是先确定每个用户的区块，这样就可以保证id和评论是一一对应的
for post in posts:
    try:
        UserID = post.find(attrs={'class':'auth'})
        Comment = post.find(attrs={'class':'postbody'})
        print("UserID: ",UserID.string)
        print("Comment: ",Comment.get_text().strip())
        print("==========================================================================")
    except:
        pass  #使用children属性会产生一些空的节点，跳过即可
browser.close()
```
